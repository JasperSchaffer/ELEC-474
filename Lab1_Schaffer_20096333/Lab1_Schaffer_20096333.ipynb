```python
import cv2
from matplotlib import pyplot as plt 
import numpy as np 
import sklearn 
print(cv2.__version__)
print(np.__version__)
print(sklearn.__version__)
```

```python
print("Hello World!")
img=cv2.imread("lena.png")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("Lena Color")
```

```python
plt.imshow(gray, cmap ='gray')
plt.title("Lena Grayscale")
```

```python
print(img.shape)
print(img.size)
print(img.dtype)
```

```python
x = 511
z=511
y=0
img2= np.zeros((512,512), np.uint8)
img3= np.zeros((512,512), np.uint8)
img4= np.zeros((512,512,3), np.uint8)
img5= np.zeros((512,512,3), np.uint8)
while z>0:
    while x>=0:
        img2[z][y]=gray[z][x]
        img3[y][z]=gray[x][z]
        img4[z][y]=img[z][x]
        img5[y][z]=img[x][z]
        y=y+1
        x=x-1
    z=z-1
    y=0
    x=511
plt.imshow(cv2.cvtColor(img2,cv2.COLOR_BGR2RGB))
```

```python
x = 511
z=511
img6= np.zeros((512,512,3), np.uint8)
img7= np.zeros((512,512), np.uint8)
while z>0:
    while x>0:
        img6[z][x][0]=255-img[z][x][0]
        img6[z][x][1]=255-img[z][x][1]
        img6[z][x][2]=255-img[z][x][2]
        img7[z][x]=255-gray[z][x]
        x=x-1
    z=z-1
    x=511
plt.imshow(cv2.cvtColor(img6,cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(img7,cv2.COLOR_BGR2RGB))
```

```python
x=511
z=511
imgb= np.zeros((512,512,3), np.uint8)
imgr= np.zeros((512,512,3), np.uint8)
imgg= np.zeros((512,512,3), np.uint8)
while z>0:
    while x>0:
        imgb[z][x][0]=img[z][x][0]
        imgg[z][x][1]=img[z][x][1]
        imgr[z][x][2]=img[z][x][2]
        x=x-1
    z=z-1
    x=511
plt.imshow(cv2.cvtColor(imgb,cv2.COLOR_BGR2RGB))
```
